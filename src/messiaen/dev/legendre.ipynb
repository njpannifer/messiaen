```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colormaps
import cmasher as cmr

from scipy.special import legendre
```

Basic code for legendre polynomials for scattering angle.

Used for partial cross sections. - see §3.2.3: R. Robinson: Fundamentals of Charged Particle Transport in Gases and Condensed Matter

```python
theta_vals = np.linspace(-np.pi, np.pi, 360 + 1)

fig, axs = plt.subplots(3, 3, figsize=(12, 12), dpi=300)
axs = axs.flatten()
for i,l in enumerate(range(9)):
    poly = legendre(l)
    r = poly(np.cos(theta_vals)) * (2*l + 1)/(4*np.pi)

    x = r*np.cos(theta_vals)
    y = r*np.sin(theta_vals)
    axs[i].plot(x,y)
plt.show()


fig = plt.figure(figsize=(12, 6), dpi=300)
ax = plt.subplot()
cs = cmr.take_cmap_colors('rainbow_r', 9, cmap_range=(0, 1), return_fmt='hex')

for i,n in enumerate(range(9)):
    poly = legendre(n)
    r = poly(np.cos(theta_vals)) * (2*l + 1)/(4*np.pi)
    ax.plot(theta_vals,r, c = cs[i], alpha = 0.75)
ax.set_xlabel(r"$\theta$", fontsize = 20)
ax.set_ylabel(r"$P_{l}(cos(\theta))$", fontsize = 20)
plt.show()
```
