```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre
```

Basic code for legendre polynomials for scattering angle.

Used for partial cross sections. - see §3.2.3: R. Robinson: Fundamentals of Charged Particle Transport in Gases and Condensed Matter

```python
theta_vals = np.linspace(-np.pi, np.pi, 360 + 1)

for n in range(6):
    poly = legendre(n)
    r = poly(np.cos(theta_vals))

    x = r*np.cos(theta_vals)
    y = r*np.sin(theta_vals)
    plt.plot(x,y)
```
